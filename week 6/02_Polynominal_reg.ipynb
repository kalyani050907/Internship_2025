```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Salary_Data.csv')
```

```python
df
```

```python
df.shape
```

```python
df.columns
```

```python
# seprate input and output
```

```python
x = df[['YearsExperience']]
```

```python
y = df['Salary']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x, y)
```

```python
lr.score(x,y)
```

```python
# Polynomial
```

```python
from sklearn.preprocessing import  PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree = 7)
```

```python
x_poly = poly.fit_transform(x)
```

```python
x_poly
```

```python
lr.fit(x_poly , y)
```

```python
lr.score(x_poly, y)
```

```python
poly = PolynomialFeatures(degree = 13)
```

```python
x_poly = poly.fit_transform(x)
```

```python
x_poly
```

```python
lr.fit(x_poly , y)
```

```python
lr.score(x_poly, y)
```

```python
plt.scatter(x,y, color='purple')
plt.plot(x , lr.predict(x_poly))
plt.grid()
```
