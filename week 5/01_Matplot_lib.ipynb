```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([12,23,32,37,44,49,51,56,59,64])
y = np.array([10,14,19,25,30,35,40,43,51,59])
z = np.array([70,65,61,58,50,48,42,37,28,15])
```

#### line plot

```python
plt.title("Line Plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(x , y, color = 'red',marker = 'o', mfc = 'blue', ls = '--');
plt.grid()

#markers => <,>,^,H,o,*
```

```python
plt.title("Line Plot")
plt.xlabel("x values")
plt.ylabel("z values")
plt.plot(x , y, color = 'purple',marker = 'H', mfc = 'blue', ls = '--');
plt.grid()
```

```python
plt.title("Line Plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(x , y, color = 'red',marker = 'o', mfc = 'blue', ls = '--');
plt.plot(x , z, color = 'purple',marker = 'H', mfc = 'blue', ls = '--');

plt.grid()
```

```python
plt.title("Line Plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(x , y, color = 'red',marker = 'o', mfc = 'blue', ls = '--',
        label = '2023');
plt.plot(x , z, color = 'purple',marker = 'H', mfc = 'blue', ls = '--',
        label = '2024');

plt.grid()
plt.legend();
```

```python
plt.figure(figsize=(16,9))
plt.subplot(2,2,1)
plt.title("Line Plot")
plt.xlabel("x values")
plt.ylabel("y values")
plt.plot(x , y, color = 'red',marker = 'o', mfc = 'blue', ls = '--');
plt.grid()

plt.subplot(2,2,2)
plt.title("Line Plot")
plt.xlabel("x values")
plt.ylabel("z values")
plt.plot(x , y, color = 'pink',marker = 'H', mfc = 'blue', ls = '--');
plt.grid()
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
###1

import matplotlib.pyplot as plt

first_10 = df.head(10)

plt.plot(first_10['PassengerId'], first_10['Fare'])  # Line plot
plt.title("Line Plot")
plt.xlabel("PassengerId")
plt.ylabel("Fare")
```

```python
###2 
pclass = df['Pclass'].value_counts().sort_index()

plt.title('Number of Passengers in Each Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.bar(pclass.index, pclass.values,color='red',edgecolor='black')
plt.grid()
```

```python
###3 

plt.figure(figsize=(16, 9))
plt.hist(df['Age'].dropna(), bins=20, color='purple', edgecolor='black')

plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.grid
```

```python

###4 

plt.scatter(df.Age,df.Fare,marker='o');
plt.xlabel('age')
plt.ylabel('fare')
plt.title('age and fare distribution')
```

```python
###5 

import pandas as pd
import matplotlib.pyplot as plt

survival_counts = df['Survived'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(survival_counts, labels=['Did Not Survive', 'Survived'], autopct='%1.1f%%')
plt.title('Survivors vs Non-Survivors');
```

```python
###6 

gender_counts = df['Sex'].value_counts()

gender_counts.plot(kind='bar')

plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.title('Passenger Count by Gender');
```

```python
###7 

gender_counts = df['Sex'].value_counts()

# Plot bar chart
gender_counts.plot(kind='bar')

# Labels and title
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.title('Passenger Count by Gender');
```

```python
###8

embarked_counts = df['Embarked'].value_counts()

plt.pie(embarked_counts, labels=embarked_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Passenger Distribution by Embarkation Port')
plt.axis('equal') 
```

```python
###9

plt.figure(figsize=(12, 5))

plt.subplot(1,2,1)
gender_counts = df['Sex'].value_counts()

gender_counts.plot(kind='bar')

plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.title('Passenger Count by Gender')

plt.subplot(1,2,2)
embarked_counts = df['Embarked'].value_counts()

plt.pie(embarked_counts, labels=embarked_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Passenger Distribution by Embarkation Port')
plt.axis('equal') 
```

```python
###10

import pandas as pd
import matplotlib.pyplot as plt

survival_counts = df.groupby(['Pclass', 'Survived']).size().unstack()

survival_counts.columns = ['Did Not Survive', 'Survived']
survival_counts.plot(kind='bar', stacked=True, color=['tomato', 'mediumseagreen'])

plt.title('Survivors and Non-Survivors by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.legend(title='Survival Status')
plt.xticks(rotation=0)
plt.tight_layout()
```
