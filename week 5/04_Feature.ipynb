```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Social_Network_Ads.csv")
df
```

```python
df.describe()
```

```python
# plot the scatterplot between age and estimated salary
```

```python
plt.title('Distribution')
plt.xlabel("EstimatedSalary	")
plt.ylabel("Age")
plt.scatter(df['EstimatedSalary'],df['Age'], c= df['Purchased']);
```

### Normalization

```python
# Seprate the data

x = df[['Age', 'EstimatedSalary']]
x
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
x_scaled
```

```python
x_scaled[:, 0].max() , x_scaled[:, 0].min() 
```

```python
x_scaled[:, 1].max() , x_scaled[:, 1].min() ,
```
